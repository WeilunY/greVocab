```python
import pymysql
import xlrd
```

```python
data = xlrd.open_workbook("./gre_vocab.xls")
```

```python
table = data.sheet_by_index(0)
```

```python
table.cell_value(1,0)
```

```python
db = pymysql.connect(host="localhost", user="root", password="123456", db="gre_vocab", autocommit=True ,charset="utf8")
```

```python
cursor = db.cursor()
```

```python
sql = """INSERT INTO vocab (id, word, chin_def) VALUES (%s, %s, %s)"""
    
sql
```

```python
for n in range(1, table.nrows):
    word = table.cell(n,0).value
    chin_def = table.cell(n,1).value
    
    values = n, word, chin_def
    
    print(values)
    
    cursor.execute(sql, values)
    
cursor.close()
db.close()
```
